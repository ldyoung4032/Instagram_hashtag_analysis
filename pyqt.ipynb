```python
## 현재 temp, rain, wind, hum, snow, cloud 받아오기

# -*- coding: utf-8 -*-
from urllib.request import urlopen, Request
import urllib
import bs4


import sys
from PyQt5.QtWidgets import *
from PyQt5.QtCore import pyqtSignal
from PyQt5.QtCore import QDate, Qt, QCoreApplication

from datetime import datetime
```

```python
class Get_Today_Data :
    
    def crawl(self) :
        enc_location = urllib.parse.quote('서울 날씨')

        url = 'https://search.naver.com/search.naver?ie=utf8&query='+ enc_location

        req = Request(url)
        page = urlopen(req)
        html = page.read()
        soup = bs4.BeautifulSoup(html,'html5lib')
        
        return soup
    
    def get_weather_data(self, soup) :
        
        temp = float(soup.find('div', class_='temperature_text').find('strong').text.replace("현재 온도", "").replace("°", ""))

        summary_list = soup.find('dl', class_ = 'summary_list')

        for idx,i in enumerate(summary_list) : ## 강수 3, 습도 5, 바람 7
            if idx==3 :
                rain = float(i.find('dd', class_ = 'desc').text.replace('mm', ''))
            elif idx==5 :
                hum = float(i.find('dd', class_ = 'desc').text.replace('%', ''))
            elif idx==7:
                wind = float(i.find('dd', class_ = 'desc').text.replace('m/s', ''))
        
        return temp, rain, hum, wind

    def __call__(self) :
        
        soup = self.crawl() ## 페이지 크롤링 정보
        temp, rain, hum, wind = self.get_weather_data(soup) ## 필요 데이터 추출
        
        return temp, rain, hum, wind
```
